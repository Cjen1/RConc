```python
from os import listdir
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
import pandas as pd
import ast # parse dictionaries from strings

%matplotlib notebook
```

```python
resultStrs = []
for res in listdir('results'):
    resultStrs.append( open('results/' + res) )

results = [ast.literal_eval(res.read()) for res in resultStrs]
```

```python
compdata = [{'ETCD Test {}'.format(str(i + 1)) : np.array(results[  2*i  ]['resps'][1:]), 
             'ZK Test {}'.format(str(i + 1)) : np.random.choice(np.array(results[2*i + 1]['resps'][1:]), len(results[  2*i  ]['resps'][1:]), replace=False)}
             for i in range(4)]
```

```python
%matplotlib notebook
dataframes = [pd.DataFrame(d) for d in compdata]


fig, axes = plt.subplots(nrows=2, ncols=2, sharey='row')

fig.suptitle("Response Times", fontsize=20)

grid = [1,1,1,1]
grid[0] = dataframes[0].boxplot(ax=axes[0,0], whis=3)
grid[1] = dataframes[1].boxplot(ax=axes[0,1], whis=3)
grid[2] = dataframes[2].boxplot(ax=axes[1,0], whis=3)
grid[3] = dataframes[3].boxplot(ax=axes[1,1], whis=3)

for p in grid:
    p.set_ylim(0, 0.175)

for p in range(2):
    grid[2*p].set_ylabel('(response times in seconds)')
```
